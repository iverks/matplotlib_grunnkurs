# Matplotlib grunnkurs

Begynn her

Matplotlib er et nyttig bibliotek for å visualisere data.

For å bruke det må vi først importere.


```python
import matplotlib.pyplot as plt
import numpy as np
```

I jupyter notebooks kan man velge mellom forskjellige konfigurasjoner av matplotlibs "magic function". Noen av disse er:

- `%matplotlib inline` -> gir bilder "inline" i notebooken
- `%matplotlib widget` -> gir interaktive plots, også inline i notebooken
- `%matplotlib qt5` -> åpner interaktive plots i et nytt vindu. Standard oppførsel for ".py"-filer, men fungerer ikke slik notebooks er lagt opp i dette faget; du må kjøre notebook-serveren på din egen pc og installere pyqt

her bruker vi `%matplotlib widget`

I dets enkleste forstand er det bare å plotte det man ønsker å plotte.


```python
ys = [0, 1, 4, 9]
plt.plot(ys)
plt.show()
```

Dersom man ikke gir inn x-verdier impliserer det at x-verdiene er 0, 1, 2, 3, 4 ... opp til n-1, der n er lengden til listen over y-verdiene.

Dermed er det følgende ekvivalent


```python
xs = [0, 1, 2, 3]
ys = [0, 1, 4, 9]
plt.plot(xs, ys)
plt.show()
```

Den standard måten å plotte på er å plotte punkter, så tegnes det automatisk streker mellom dem. Man kan enkelt plotte uten disse strekene ved å sende inn en formatteringsstreng som parameter.

Merk at når man plotter på denne måten kommer plottene oppå hverandre. For å blanke ut figuren kan vi bruke `plt.clf()` som står for "clear figure". Kommenter ut linje 4 for å se hvordan det ser ut dersom man ikke gjør det.

Andre måter å ikke la plots overlappe kommer i neste notebook.


```python
xs = [0, 1, 2, 3]
ys = [0, 1, 4, 9]
plt.plot(xs, ys)
plt.clf()
plt.plot(xs, ys, ".")  # eller "o" for større prikker
plt.show()
```

## Å legge ved annen informasjon

For å legge ved annen informasjon kan man bruke diverse funksjoner.

- `plt.xlabel()` og `plt.ylabel()` navngir aksene
- `plt.title()` titulerer plottet
- `plt.legend()` lager den lille boksen oppe i hjørnet som forteller oss hva hver graf betyr. Den henter navnet fra `label` som vi gir inn som et argument til `plt.plot()`


```python
xs = [0, 1, 2, 3]
ys = [0, 1, 4, 9]
ys2 = [0, 0, 0, 0]
plt.clf()
plt.plot(xs, ys, label="studenter")  # eller "o" for større prikker
plt.plot(xs, ys2, label="studentassistenter")  # eller "o" for større prikker
plt.xlabel("Timer lagt inn i itgk-øvinger i uken [t]")
plt.ylabel("Relativ økning i lønn [%]")
plt.title("Analyse av lønn for forskjellige arbeidsinnsatser i itgk")
plt.legend()
plt.show()
```

Dersom man har lister med veldig mange, eller noen veldig store verdier kan man også fikse det:

- `plt.ylim()` begrenser plottet i y-retning, som kan være hjelpsomt dersom man har en funksjon som divergerer
- `plt.xlim()` begrenser plottet i x-retning, som kan være hjelpsomt for å zoome inn på en interessant del
- `plt.xticks()` og `plt.yticks()` bestemmer hvilke nummer på aksene som skal være med. Merk at disse overskriver de to forrige funksjonene dersom de kalles på sist. **Test dette i notebooken**


```python
xs = [n for n in range(365)]
ys = [n % 7 for n in range(365)]
ys[100] *= 100
plt.clf()
plt.plot(xs, ys, label="studenter")  # eller "o" for større prikker
plt.title("humøret mitt per dag i året")
plt.xticks([n for n in range(0, 365, 7)] + [95])
plt.yticks([n for n in range(10 + 1) if n not in [2, 7]])
plt.xlim([50, 125])
plt.ylim([0, 10])
plt.show()
```

## Man kan også plotte på andre vis

- `plt.scatter()` gir scatter plots som viser korrelasjon mellom to egenskaper
- `plt.pie()` gir pieplots som viser andelsfordeling
- `plt.bar()` gir barcharts som viser mengder som funksjon av en diskret variabel
- `plt.hist()` gir histogrammer som viser mengder som funksjon av en kontinuerlig variabel. Merk at histogrammer er litt spesielle fordi du kan sende inn som parameter hvor mange kolonner du ønsker. Standard er 10.


```python
x = np.linspace(0, 250, 1000)
skills = np.random.normal(loc=1+(1 - (((x - 170) / 175)**2))*8, scale=(1 - (((x - 150) / 175)**2))* 1)
plt.clf()
plt.scatter(x, skills)
plt.xlabel("Finn på enhet")
plt.ylabel("Matplotlib-skills fra 1 til 10")
plt.show()
```

```python
plt.clf()
plt.pie([2, 6, 4], labels=["a", "b", "c"])
plt.show()
```

```python
plt.clf()
plt.bar(["a", "b", "c", "d"],[1, 5, 3, 6])
plt.show()
```

```python
samples = np.random.normal(loc=1, size=1000)
plt.clf()
# plt.hist(samples, 50)
plt.hist(samples)
plt.show()
```
