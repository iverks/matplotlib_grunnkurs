# 2D data

Når du har flerdimensjonelle data trenger du en ny måte å plotte det på. 

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget
```

Her kommer `plt.imshow()`

```python
x = np.linspace(0, 50, 1000)
y = np.linspace(0, 50, 1000)
x, y = np.meshgrid(x, y)

center = (20, 40)
cx, cy = center

r = np.sqrt((x - cx) ** 2 + (y - cy) ** 2)
z = np.sin(r)

fig, ax = plt.subplots()
ax.imshow(z)
```

## Vektorfelt

For å plotte vektorfelt kan du bruke metoden `plt.quiver()`. Det plottes på formen `plt.quiver(x, y, u, v)` dersom vektorfeltet er:

$$
\vec{F}(x, y) = u(x, y)\hat{x} + v(x, y)\hat{y}
$$

I eksempelet under har vi at

$$
\vec{F}(x, y) = sin(x+y)\hat{x} + sin(x-y)\hat{y}
$$

```python
x = np.linspace(0, 5, 20)
y = np.linspace(0, 5, 20)
x, y = np.meshgrid(x, y)

u = np.sin(x + y)
v = np.sin(x - y)

fig, ax = plt.subplots()
ax.quiver(x, y, u, v)
```

Det går også an å farge pilene vilkårlig. Dette kan f.eks. være å farge dem ved hjelp av fargehjulet ut i fra retningen de peker (`np.arctan2(u, v), cmap="hsv"`). Dersom vektoren har en z-komponent kan pilene farges ut i fra den. Dette er vist i eksempelet under. Der antar vi at vektorfeltet har formen

$$
\vec{F}(x, y) = sin(x+y)\hat{x} + sin(x-y)\hat{y} + cos(x)\hat{z}
$$


```python
x = np.linspace(0, 5, 20)
y = np.linspace(0, 5, 20)
x, y = np.meshgrid(x, y)

u = np.sin(x + y)
v = np.sin(x - y)
z = np.cos(x)

fig, ax = plt.subplots()
cax = ax.quiver(x, y, u, v, z, cmap="magma")
fig.colorbar(cax)
```

## Du kan også plotte data i 3d

Men selv om dette er veldig kult er det dessverre ikke alltid like nyttig

```python
x = np.linspace(0, 50, 1000)
y = np.linspace(0, 50, 1000)
x, y = np.meshgrid(x, y)

center = (25, 25)
cx, cy = center

r = np.sqrt((x - cx) ** 2 - (y - cy) ** 2)
z = np.sin(r)


fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, z, cmap="magma")
```
