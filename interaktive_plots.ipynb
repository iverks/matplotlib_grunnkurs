# Interaktive plots

Dette er moro, og kan være nyttig for å forstå hvordan en funksjon avhenger av en variabel. Det er ikke pensum

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
import numpy as np

%matplotlib widget
```

```python
fig, ax = plt.subplots()
xs = np.linspace(0, 2*np.pi)

@widgets.interact(theta = (0, 10, 0.05), k = widgets.fixed(1))
def plot_cos(theta, k):
    """Husk at det er en fordel å gjøre så lite som mulig inne i denne funksjonen siden den kjøres veldig mange ganger."""
    ys = np.cos(k*xs + theta)
    [l.remove() for l in ax.lines]
    ax.plot(xs, ys, color="C3")
```
